# Boolean indexing

```python
D = np.random.randint(0,10,[5,5])
D
```

```python
# Masque
D < 5
```

```python
# Convertir les valeurs de tableau >< seuil
D[D<5]=10
D
```

```python
E = np.random.randint(0,10,[5,5])
E
```

```python
E[(E<5) & (E>2)]=10
E
```

```python
plt.imshow(E)
plt.axis('off')
plt.show()
```

```python
from scipy import misc
face = misc.face()
plt.imshow(face)
plt.show()
```

```python
face.shape
```

```python
face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
face.shape
```

```python
N = face[:600,:900]
# N[N<50]=0
# N[N>200] = 255
plt.imshow(N, cmap=plt.cm.gray )
plt.show()
face.shape
```

```python
# Zoom de 1/4 vers le centre de cette image
h = face.shape[0]
w = face.shape[1]
image = face[h//4 : -h//4,w//4 : -w//4]
# image[image>200]=255
plt.imshow(image, cmap=plt.cm.gray )
plt.show()
face.shape
```

```python
# Résuire de moitier de poids de cette images sans perdre son qualité 
face_reduit =  face[::2, ::2]
plt.imshow(face_reduit, cmap=plt.cm.gray )
plt.show()
face.shape
```

```python
# Résuire de moitier de poids de cette images sans perdre son qualité 
face_reduit1 =  face_reduit[::2, ::2]
plt.imshow(face_reduit1, cmap=plt.cm.gray )
plt.show()
face.shape
```

```python
face_reduit2 =  face_reduit1[::2, ::2]
plt.imshow(face_reduit2, cmap=plt.cm.gray )
plt.show()
face.shape
```
